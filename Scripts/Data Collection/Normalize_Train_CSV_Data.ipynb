```python
import pandas as pd
import math
import numpy as np
```

```python
def angle_between_landmarks(landmark1, landmark2, landmark3):
    vector_AB = (landmark1[0] - landmark2[0], landmark1[1] - landmark2[1], landmark1[2] - landmark2[2])
    vector_BC = (landmark3[0] - landmark2[0], landmark3[1] - landmark2[1], landmark3[2] - landmark2[2]) 
    dot_product = np.dot(vector_AB, vector_BC)
    magnitude1 = np.linalg.norm(vector_AB)
    magnitude2 = np.linalg.norm(vector_BC)
    cosine_angle = dot_product / (magnitude1 * magnitude2)
    if cosine_angle < -1:
        cosine_angle = -1
    elif cosine_angle > 1:
        cosine_angle = 1
    angle = math.acos(cosine_angle)
    angle_degrees = np.degrees(angle)
    return angle_degrees
```

```python
df = pd.read_csv('D:\Chord_Data\Ring\\Ring_test_data.csv')
```

```python
df
```

```python
n = 4 #number of landmarks
for i in range(len(df)):
    x_pos, y_pos, z_pos = df.iloc[i]['Landmark_1_x'], df.iloc[i]['Landmark_1_y'], df.iloc[i]['Landmark_1_z']
    df.iloc[i, 1:n+1] -= x_pos
    df.iloc[i, n+1:2*n+1] -= y_pos
    df.iloc[i, 2*n+1:] -= z_pos
    df.iloc[i, 1:] *= 1/(df.iloc[i, 1:n+1].max() - df.iloc[i, 1:n+1].min()) #normalize hand for chord
    df.iloc[i, 1:] *= 1/(df.iloc[i, n+1:2*n+1].max() - df.iloc[i, n+1:2*n+1].min()) #normalize for finger
    df.iloc[i, 1:] *= 1/(df.iloc[i, 3:5].max() - df.iloc[i, 3:5].min()) #normalize for wrist angle
    
```

```python
df
```

```python
for i in range(4):
    df[f'Coordinates_{i+1}'] = df.apply(lambda row: [row[f'Landmark_{i+1}_x'], row[f'Landmark_{i+1}_y'], row[f'Landmark_{i+1}_z']], axis=1)
```

```python
df
```

```python
number_list = [i for i in range(1, n+1)]
for i in range(n):
    index_list = number_list.copy()
    index_list.pop(i)
    print(index_list)
    new_column_name = f'Angle_{i+1}'
    lm1 = df[f'Coordinates_{index_list[0]}']
    lm2 = df[f'Coordinates_{index_list[1]}']
    lm3 = df[f'Coordinates_{index_list[2]}']
    df[new_column_name] = lm1
    for j in range(len(lm1)):
        df[new_column_name][j] = angle_between_landmarks(lm1[j], lm2[j], lm3[j])
```

```python
for i in range(n):
    df = df.drop(columns=[f'Coordinates_{i+1}'])
```
